# แก้ label

```python
import os
import glob

label_dir = 'fish_data/valid/labels'  # path label dataset ใหม่

for label_file in glob.glob(f'{label_dir}/**/*.txt', recursive=True):
    with open(label_file, 'r') as f:
        lines = f.readlines()
    
    new_lines = []
    for line in lines:
        parts = line.strip().split()
        parts[0] = '0'  # เปลี่ยนทุก class เป็น 0 (fish)
        new_lines.append(' '.join(parts) + '\n')
    
    with open(label_file, 'w') as f:
        f.writelines(new_lines)

print('แก้ label เสร็จแล้ว!')
```

# เช็ค

```python
import glob

def check_classes(label_dir):
    classes = set()
    total_boxes = 0
    
    for label_file in glob.glob(f'{label_dir}/**/*.txt', recursive=True):
        with open(label_file, 'r') as f:
            for line in f:
                parts = line.strip().split()
                if parts:
                    classes.add(parts[0])
                    total_boxes += 1
    
    print(f'จำนวน class: {len(classes)}')
    print(f'class IDs: {sorted(classes)}')
    print(f'total boxes: {total_boxes}')

# เช็ค dataset ใหม่
check_classes('fish_data/valid/labels')
```
